# Object-Oriented Programming: Polymorphism (Python)

Polymorphism is a core concept of **Object-Oriented Programming (OOP)** that allows **one method name to have multiple behaviors** depending on the object calling it.

---

### What is Polymorphism?

**Definition:**  
Polymorphism means **one interface, many implementations**.

Breakdown of the word:
- **Poly** → Many  
- **Morphism** → Forms  

So, the concept implies:

> A single function or method can behave differently for different objects.

---

### Real-Life Example

Consider the method:

```
make_sound()
```

Different animals respond differently:

| Animal | Behavior  |
|--------|-----------|
| Dog    | Bark      |
| Cat    | Meow      |
| Cow    | Moo       |

Here, **`make_sound()`** is the same method name, but each animal implements it differently.

**Key Point:**  
- **Same function name → different behavior**  
- This allows for **flexible and reusable code** in OOP.

---

### Question: Animal Sound System (Using Polymorphism)

#### Problem Description
Create a system where **different animals make different sounds**.

All animals share:
- `name`

But each animal produces a **different sound**.

**Example**
- Dog → Bark  
- Cat → Meow  
- Cow → Moo

We will use **polymorphism** so the **same method name works differently** for each animal.

---

#### Requirements

**Parent Class → `Animal`**

**Attributes**
- `name`

**Method**
- `make_sound()` → This will be **overridden** by child classes

---

**Child Classes**
- `Dog`
- `Cat`
- `Cow`

Each child class should **override the `make_sound()` method** to produce its own sound.

---

#### Concepts Tested
- Inheritance  
- Method overriding  
- Polymorphism  
- Object-specific behavior

````python
class Animal:

    def __init__(self, name):
        self.name = name

    def make_sound(self):
        print("Animal makes a sound")


class Dog(Animal):

    def make_sound(self):
        print(self.name, "says: Bark")


class Cat(Animal):

    def make_sound(self):
        print(self.name, "says: Meow")


class Cow(Animal):

    def make_sound(self):
        print(self.name, "says: Moo")


animals = [
    Dog("Tommy"),
    Cat("Kitty"),
    Cow("Ganga")
]

for animal in animals:
    animal.make_sound()
````

### Question: Shape Area Calculator (Using Polymorphism)

#### Problem Description
Create a system that **calculates the area of different shapes**.

Shapes include:
- Rectangle  
- Circle  
- Triangle  

Each shape calculates its area differently, but we will use **the same method name** for all.

**Method to use:**  
```
calculate_area()
```

---

#### Requirements

**Parent Class → `Shape`**

**Method**
- `calculate_area()` → To be overridden by child classes

---

**Child Classes**
- `Rectangle`  
- `Circle`  
- `Triangle`

Each child class should **override `calculate_area()`** to calculate its own area based on shape-specific formulas.

---

#### Concepts Tested
- Polymorphism  
- Method overriding  
- Reusing method names for multiple behaviors

````python
import math


class Shape:

    def calculate_area(self):
        print("Area calculation")


class Rectangle(Shape):

    def __init__(self, length, width):
        self.length = length
        self.width = width

    def calculate_area(self):
        area = self.length * self.width
        print("Rectangle Area:", area)


class Circle(Shape):

    def __init__(self, radius):
        self.radius = radius

    def calculate_area(self):
        area = math.pi * self.radius * self.radius
        print("Circle Area:", area)


class Triangle(Shape):

    def __init__(self, base, height):
        self.base = base
        self.height = height

    def calculate_area(self):
        area = 0.5 * self.base * self.height
        print("Triangle Area:", area)


shapes = [
    Rectangle(10, 5),
    Circle(7),
    Triangle(6, 8)
]

for shape in shapes:
    shape.calculate_area()
````

### Question: Payment System (Using Polymorphism)

#### Problem Description
Create a **payment system** that supports different types of payments.

Payment methods include:
- CreditCard  
- UPI  
- PayPal  

Each payment method processes payments **differently**, but we use **one common method name**:

```
pay(amount)
```

---

#### Requirements

**Parent Class → `Payment`**

**Method**
- `pay(amount)` → To be implemented differently by each child class

---

**Child Classes**
- `CreditCard`  
- `UPI`  
- `PayPal`

Each child class should **implement its own payment logic** while using the same method name `pay(amount)`.

---

#### Concepts Tested
- Polymorphism  
- Method overriding  
- Consistent interface with multiple behaviors  
- Object-specific functionality

### Visual Understanding of Polymorphism

```
            Payment
               |
   ---------------------------
   |            |            |
CreditCard     UPI        PayPal
   |            |            |
  pay()        pay()        pay()
```

**Key Idea:**  
- Same method name → `pay()`  
- Different behavior depending on the object (CreditCard, UPI, PayPal)  

This demonstrates **polymorphism** in action: **one interface, many implementations**.

````python
class Payment:

    def pay(self, amount):
        print("Processing payment")


class CreditCard(Payment):

    def pay(self, amount):
        print("Paid", amount, "using Credit Card")


class UPI(Payment):

    def pay(self, amount):
        print("Paid", amount, "using UPI")


class PayPal(Payment):

    def pay(self, amount):
        print("Paid", amount, "using PayPal")


payments = [
    CreditCard(),
    UPI(),
    PayPal()
]

for payment in payments:
    payment.pay(1000)
````
