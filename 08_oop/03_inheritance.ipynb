# Object-Oriented Programming: Inheritance (Python)

In this notebook, we will learn the next important concept in **Object-Oriented Programming (OOP)**:

- **Inheritance**

Inheritance allows one class to **reuse the properties and methods of another class**, making code more organized and reusable.

---

## 1️⃣ Understanding Inheritance (Concept)

**Inheritance means:**

A class can reuse the **attributes and methods** of another class.

Think of a **real-life example**:

```
Animal
 ├── Dog
 ├── Cat
 └── Bird
```

All animals have common behaviors such as:
- `name`
- `eat()`
- `sleep()`

Instead of writing these properties again for **Dog, Cat, and Bird**, we create a **base class (`Animal`)** and let the other classes inherit from it.

Example concept:

```
class Dog(Animal):
```

This means:

- `Dog` **inherits everything from** `Animal`.

---

## 2️⃣ Simple Example of Inheritance

```python
class Animal:

    def eat(self):
        print("Animal is eating")


class Dog(Animal):

    def bark(self):
        print("Dog is barking")


dog1 = Dog()

dog1.eat()
dog1.bark()
```

**Output**

```
Animal is eating
Dog is barking
```

---

## Why Does This Work?

Because **`Dog` inherits from `Animal`**, the `Dog` class automatically gets access to:

- `eat()` method from `Animal`
- Its own method `bark()`

This allows us to **reuse code instead of rewriting the same methods in multiple classes.**

### Question: Employee Management System (Using Inheritance)

#### Problem Description
You are designing a **company employee system**.

All employees share some **common data**:
- `name`
- `id`

However, their **salary calculation differs** depending on the type of employee.

**Employee Types**
- Manager
- Developer

Using **inheritance**, we can **reuse common attributes and methods** while customizing specific behaviors.

---

#### Requirements

**Parent Class → `Employee`**

**Attributes**
- `name`
- `emp_id`

**Method**
- `display_info()` → Displays employee details

---

**Child Class → `Manager`**

**Additional Attribute**
- `bonus`

**Method**
- `calculate_salary()` → Calculates manager’s salary including bonus

---

**Child Class → `Developer`**

**Additional Attributes**
- `hourly_rate`
- `hours_worked`

**Method**
- `calculate_salary()` → Calculates developer’s salary based on hours worked

---

#### Concepts Tested
- Inheritance  
- Constructor reuse  
- Child class attributes  
- Creating and using multiple objects

````python
class Employee:

    def __init__(self, name, emp_id):
        self.name = name
        self.emp_id = emp_id

    def display_info(self):
        print("Employee Name:", self.name)
        print("Employee ID:", self.emp_id)


class Manager(Employee):

    def __init__(self, name, emp_id, bonus):
        super().__init__(name, emp_id)
        self.bonus = bonus

    def calculate_salary(self):
        salary = 50000 + self.bonus
        print("Manager Salary:", salary)


class Developer(Employee):

    def __init__(self, name, emp_id, hourly_rate, hours):
        super().__init__(name, emp_id)
        self.hourly_rate = hourly_rate
        self.hours = hours

    def calculate_salary(self):
        salary = self.hourly_rate * self.hours
        print("Developer Salary:", salary)


m1 = Manager("Vinod", 101, 10000)
d1 = Developer("Bhavi", 102, 500, 80)

m1.display_info()
m1.calculate_salary()

print("------------------")

d1.display_info()
d1.calculate_salary()
````

### Question: Animal Sound System (Using Inheritance)

#### Problem Description
Create a system where **different animals make different sounds**.

All animals share:
- `name`

But each animal makes a **different sound**.

**Example**
- Dog → Bark  
- Cat → Meow  
- Cow → Moo

---

#### Requirements

**Parent Class → `Animal`**

**Attributes**
- `name`

**Method**
- `display_name()` → Displays the animal’s name

---

**Child Classes**
- `Dog`  
- `Cat`  
- `Cow`  

Each child class should have a method:
- `make_sound()` → Prints the sound of the animal

````python
class Animal:

    def __init__(self, name):
        self.name = name

    def display_name(self):
        print("Animal:", self.name)


class Dog(Animal):

    def make_sound(self):
        print("Bark")


class Cat(Animal):

    def make_sound(self):
        print("Meow")


class Cow(Animal):

    def make_sound(self):
        print("Moo")


dog = Dog("Tommy")
cat = Cat("Kitty")
cow = Cow("Ganga")

dog.display_name()
dog.make_sound()

cat.display_name()
cat.make_sound()

cow.display_name()
cow.make_sound()
````
