# Object-Oriented Programming: Encapsulation (Python)

Encapsulation is a fundamental OOP concept that **hides internal data** and **controls how it is accessed**.

---

### Concept

Think of encapsulation like a **capsule**:

- Users **cannot directly change important data**.  
- They must interact with the data using **methods**.  

**Example: Bank Account**

You should **not directly change the balance**:

```python
account.balance = -100000  # ❌ Not allowed
```

Instead, use controlled methods:

```python
account.deposit(5000)
account.withdraw(1000)
```

**Encapsulation Advantages**
- Protects data  
- Controls access  
- Prevents mistakes  

---

### Access Modifiers in Python

Python uses **naming conventions** to indicate access levels:

| Modifier      | Convention      | Description                   |
|---------------|----------------|-------------------------------|
| Public        | `variable`      | Accessible anywhere           |
| Protected     | `_variable`     | Accessible within class & subclass |
| Private       | `__variable`    | Accessible only within class  |

**Example:**

```python
self.__balance  # Private variable
```

The double underscore `__` makes the attribute private and prevents direct modification from outside the class.

### Question 1: Secure Bank Account System (Using Encapsulation)

#### Problem Description
Create a **bank account system** where the **balance cannot be directly modified**.

The balance should be **private**, and users must interact with it using methods:

- Deposit money  
- Withdraw money  
- Check balance

---

#### Requirements

**Class → `BankAccount`**

**Attributes**
- `account_holder`  
- `__balance` (private)

**Methods**
- `deposit(amount)` → Adds money to the account  
- `withdraw(amount)` → Withdraws money (cannot exceed balance)  
- `check_balance()` → Displays the current balance

---

#### Rules
- Users **cannot withdraw more than the available balance**  
- Balance must **only be modified via methods**, not directly

````python
class BankAccount:

    def __init__(self, holder, balance):
        self.holder = holder
        self.__balance = balance   # private variable

    def deposit(self, amount):
        self.__balance += amount
        print("Deposited:", amount)

    def withdraw(self, amount):
        if amount > self.__balance:
            print("Insufficient balance")
        else:
            self.__balance -= amount
            print("Withdrawn:", amount)

    def check_balance(self):
        print("Current Balance:", self.__balance)


account = BankAccount("Vinod", 10000)

account.deposit(2000)
account.withdraw(3000)
account.check_balance()
````

### Question 2: Student Marks System (Using Encapsulation)

#### Problem Description
Create a **student system** where **marks are private** and can only be updated or accessed through methods.

---

#### Requirements

**Class → `Student`**

**Attributes**
- `name`  
- `__marks` (private)

**Methods**
- `set_marks(marks)` → Updates the student's marks  
- `get_marks()` → Returns the current marks  
- `display_student()` → Displays student details

````python
class Student:

    def __init__(self, name, marks):
        self.name = name
        self.__marks = marks

    def set_marks(self, marks):
        if marks >= 0 and marks <= 100:
            self.__marks = marks
        else:
            print("Invalid marks")

    def get_marks(self):
        return self.__marks

    def display_student(self):
        print("Name:", self.name)
        print("Marks:", self.__marks)


s1 = Student("Vinod", 80)

s1.display_student()

s1.set_marks(90)

print("Updated Marks:", s1.get_marks())
````
