# 1. Student Information System

## Problem Description
Create a system to store and display student information.

Each student should have:
- **name**
- **age**
- **course**

The program should allow creating multiple students and printing their details.

---

## Requirements

### Create a class called `Student`

#### Attributes
- `name`
- `age`
- `course`

#### Methods
- `display_info()` → prints student details

---

## Concepts Tested
- Class creation
- Object creation
- Accessing attributes
- Calling methods

```python
class Student:

    def __init__(self, name, age, course):
        self.name = name
        self.age = age
        self.course = course

    def display_info(self):
        print("Name:", self.name)
        print("Age:", self.age)
        print("Course:", self.course)
        print("-------------------")


# Creating objects
student1 = Student("Vinod", 21, "Data Science")
student2 = Student("Rahul", 22, "Computer Science")

# Calling method
student1.display_info()
student2.display_info()
```
