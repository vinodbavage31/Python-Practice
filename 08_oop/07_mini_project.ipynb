# Mini Project: Personal Habit Tracker System

This project is a **self-improvement habit tracking application** where users can create and track daily habits. The system is designed using **Object-Oriented Programming principles**.

---

## Project Overview

The system allows users to:

1. Register a new user  
2. Create habits  
3. View all habits  
4. Mark habits as completed  
5. View habit progress  
6. Delete habits  
7. Exit the program  

The program should **run continuously in a loop** until the user chooses to exit.

---

## System Design

The system is designed with **multiple classes** to ensure modularity, encapsulation, and maintainability.

---

### 1️⃣ Class: `Habit`

Represents a habit that a user wants to track.

**Attributes**
- `habit_name` → Name of the habit (e.g., "Workout")  
- `category` → Habit category (e.g., "Health")  
- `completed_days` → Number of days the habit has been completed  
- `total_days` → Total days tracked  

**Methods**
- `mark_completed()` → Increments `completed_days` and `total_days`  
- `show_progress()` → Displays current progress  
- `display_habit()` → Prints habit details  

**Example Output**
```
Workout | Health | Progress: 5/7
```

---

### 2️⃣ Abstract Class: `User`

Defines a **user interface** using abstraction.

**Attributes**
- `name`  
- `user_id`  

**Abstract Methods**
- `show_role()` → Defines the role of the user

---

### 3️⃣ Class: `HabitUser` (Inherits `User`)

Represents a user who tracks habits.

**Attributes**
- `name`  
- `user_id`  
- `__habits` → Private list of `Habit` objects  

**Methods**
- `add_habit()` → Adds a new habit to the user  
- `view_habits()` → Displays all habits  
- `mark_habit_complete()` → Marks a habit as completed  
- `delete_habit()` → Removes a habit  
- `view_progress()` → Displays progress for each habit  
- `show_role()` → Returns the role of the user  

**Encapsulation:**  
The `__habits` attribute is private and accessible only through methods.

**Polymorphism:**  
The `show_role()` method can be overridden by different user types (e.g., Admin).

---

### 4️⃣ Class: `HabitTracker`

Manages the overall system.

**Attributes**
- `users` → Dictionary to store users  
```python
users = {
 "101": user_object,
 "102": user_object
}
```

**Methods**
- `register_user()` → Adds a new `HabitUser`  
- `find_user()` → Retrieves a user by ID  
- `display_users()` → Shows all registered users

---

## Program Flow

When the program starts:

```
Welcome to Habit Tracker
```

**Menu:**

1. Register User  
2. Create Habit  
3. View All Habits  
4. Mark Habit Completed  
5. View Progress  
6. Delete Habit  
7. Exit  

The program executes the selected feature and continues until the user exits.

### Feature Details

**1. Register User**
- Inputs: Name, User ID  
- Action: Creates a `HabitUser` object and stores it in the `users` dictionary

**2. Create Habit**
- Inputs: User ID, Habit Name, Category  
- Action: Creates a `Habit` object and adds it to the user's habit list

**3. View All Habits**
- Inputs: User ID  
- Action: Displays all habits for the user

**4. Mark Habit Completed**
- Inputs: User ID, Habit Name  
- Action: Increments `completed_days` and `total_days` for the selected habit

**5. View Progress**
- Displays:
```
Workout → 5 / 7 days
Reading → 3 / 7 days
Meditation → 2 / 7 days
```
- Optionally calculate success rate:
```
success_rate = (completed_days / total_days) * 100
```

**6. Delete Habit**
- Inputs: Habit Name  
- Action: Removes the habit from the user's list

---

## Data Structures Used

- **Dictionary** → To store users (`users`)  
- **List** → To store habits for each user  
- **Objects** → Each habit is an object

````python
from abc import ABC, abstractmethod
from datetime import date


class Habit:
    def __init__(self, habit_name: str, category: str):
        self.habit_name = habit_name
        self.category = category
        self.completed_days = 0
        self.total_days = 0
        self.streak = 0
        self._last_completed: date | None = None

    def mark_completed(self):
        today = date.today()

        if self._last_completed == today:
            print(f"'{self.habit_name}' already marked complete today.")
            return

        self.completed_days += 1
        self.total_days += 1

        if self._last_completed:
            delta = (today - self._last_completed).days
            self.streak = self.streak + 1 if delta == 1 else 1
        else:
            self.streak = 1

        self._last_completed = today
        print(f"'{self.habit_name}' marked complete. Streak: {self.streak}")

    def show_progress(self):
        if self.total_days == 0:
            print(f"{self.habit_name}: No progress yet.")
            return

        rate = (self.completed_days / self.total_days) * 100
        print(
            f"{self.habit_name} ({self.category}) "
            f"{self.completed_days}/{self.total_days} days "
            f"{rate:.1f}% streak:{self.streak}"
        )

    def display_habit(self):
        print(
            f"{self.habit_name} | {self.category} | "
            f"{self.completed_days}/{self.total_days}"
        )

    def __str__(self):
        return (
            f"{self.habit_name} | {self.category} | "
            f"{self.completed_days}/{self.total_days}"
        )


class User(ABC):
    def __init__(self, name: str, user_id: str):
        self.name = name
        self.user_id = user_id

    @abstractmethod
    def show_role(self):
        pass

    def __str__(self):
        return f"{self.name} ({self.user_id})"


class HabitUser(User):
    def __init__(self, name: str, user_id: str):
        super().__init__(name, user_id)
        self.__habits: list[Habit] = []

    def show_role(self):
        return "Habit Tracker User"

    def add_habit(self, habit_name: str, category: str):
        for h in self.__habits:
            if h.habit_name.lower() == habit_name.lower():
                print("Habit already exists.")
                return

        self.__habits.append(Habit(habit_name, category))
        print("Habit added.")

    def view_habits(self):
        if not self.__habits:
            print("No habits found.")
            return

        for h in self.__habits:
            h.display_habit()

    def mark_habit_complete(self, habit_name: str):
        habit = self._find_habit(habit_name)

        if habit:
            habit.mark_completed()
        else:
            print("Habit not found.")

    def delete_habit(self, habit_name: str):
        habit = self._find_habit(habit_name)

        if habit:
            self.__habits.remove(habit)
            print("Habit deleted.")
        else:
            print("Habit not found.")

    def view_progress(self):
        if not self.__habits:
            print("No habits to show.")
            return

        for h in self.__habits:
            h.show_progress()

    def weekly_report(self):
        if not self.__habits:
            print("No habits available.")
            return

        for h in self.__habits:
            print(f"{h.habit_name} ({h.category}) {h.completed_days}/7")

    def _find_habit(self, habit_name: str):
        for h in self.__habits:
            if h.habit_name.lower() == habit_name.lower():
                return h
        return None


class HabitTracker:
    def __init__(self):
        self.users: dict[str, HabitUser] = {}

    def register_user(self, name: str, user_id: str):
        if user_id in self.users:
            print("User ID already exists.")
            return

        self.users[user_id] = HabitUser(name, user_id)
        print("User registered.")

    def find_user(self, user_id: str):
        return self.users.get(user_id)

    def display_users(self):
        if not self.users:
            print("No users registered.")
            return

        for user in self.users.values():
            print(user)


def get_user(tracker: HabitTracker):
    uid = input("Enter user ID: ").strip()
    user = tracker.find_user(uid)

    if not user:
        print("User not found.")

    return user


def print_menu():
    print("\nHabit Tracker")
    print("1 Register User")
    print("2 Create Habit")
    print("3 View All Habits")
    print("4 Mark Habit Completed")
    print("5 View Progress")
    print("6 Delete Habit")
    print("7 Weekly Report")
    print("8 Show All Users")
    print("9 Exit")


def main():
    tracker = HabitTracker()

    while True:
        print_menu()
        choice = input("Select option: ").strip()

        if choice == "1":
            name = input("Enter name: ").strip()
            uid = input("Enter user ID: ").strip()
            tracker.register_user(name, uid)

        elif choice == "2":
            user = get_user(tracker)
            if user:
                habit_name = input("Habit name: ").strip()
                category = input("Category: ").strip()
                user.add_habit(habit_name, category)

        elif choice == "3":
            user = get_user(tracker)
            if user:
                user.view_habits()

        elif choice == "4":
            user = get_user(tracker)
            if user:
                habit_name = input("Habit name: ").strip()
                user.mark_habit_complete(habit_name)

        elif choice == "5":
            user = get_user(tracker)
            if user:
                user.view_progress()

        elif choice == "6":
            user = get_user(tracker)
            if user:
                habit_name = input("Habit name: ").strip()
                user.delete_habit(habit_name)

        elif choice == "7":
            user = get_user(tracker)
            if user:
                user.weekly_report()

        elif choice == "8":
            tracker.display_users()

        elif choice == "9":
            print("Exiting program.")
            break

        else:
            print("Invalid option.")


if __name__ == "__main__":
    main()
````
