### Step 1 — Book Class

#### Problem Description
Create a class to represent a **book** in the library.  

Each book should contain:
- `title`  
- `author`  
- `availability`  

#### Code
```python
class Book:

    def __init__(self, title, author):
        self.title = title
        self.author = author
        self.available = True

    def display_info(self):
        status = "Available" if self.available else "Borrowed"
        print(f"{self.title} by {self.author} - {status}")
```

---

### Step 2 — Abstract User Class

We will use **abstraction** here.  
All users must implement:
- `borrow_book()`  
- `return_book()`  

#### Code
```python
from abc import ABC, abstractmethod

class User(ABC):

    def __init__(self, name):
        self.name = name

    @abstractmethod
    def borrow_book(self, book):
        pass

    @abstractmethod
    def return_book(self, book):
        pass
```

---

### Step 3 — Student Class

`Student` inherits from `User`.

#### Code
```python
class Student(User):

    def borrow_book(self, book):
        if book.available:
            book.available = False
            print(self.name, "borrowed", book.title)
        else:
            print("Book not available")

    def return_book(self, book):
        book.available = True
        print(self.name, "returned", book.title)
```

---

### Step 4 — Teacher Class

`Teacher` is also a type of user.

#### Code
```python
class Teacher(User):

    def borrow_book(self, book):
        if book.available:
            book.available = False
            print(self.name, "borrowed", book.title)
        else:
            print("Book not available")

    def return_book(self, book):
        book.available = True
        print(self.name, "returned", book.title)
```

> **Note:**  
> Both `Student` and `Teacher` implement the same methods — this demonstrates **polymorphism**.

### Step 5 — Library Class

The library **manages the collection of books**.

#### Code
```python
class Library:

    def __init__(self):
        self.books = []

    def add_book(self, book):
        self.books.append(book)
```

````python
from abc import ABC, abstractmethod


class Book:

    def __init__(self, title, author):
        self.title = title
        self.author = author
        self.available = True

    def display_info(self):
        status = "Available" if self.available else "Borrowed"
        print(f"{self.title} by {self.author} - {status}")


class User(ABC):

    def __init__(self, name):
        self.name = name

    @abstractmethod
    def borrow_book(self, book):
        pass

    @abstractmethod
    def return_book(self, book):
        pass


class Student(User):

    def borrow_book(self, book):
        if book.available:
            book.available = False
            print(self.name, "borrowed", book.title)
        else:
            print("Book not available")

    def return_book(self, book):
        book.available = True
        print(self.name, "returned", book.title)


class Teacher(User):

    def borrow_book(self, book):
        if book.available:
            book.available = False
            print(self.name, "borrowed", book.title)
        else:
            print("Book not available")

    def return_book(self, book):
        book.available = True
        print(self.name, "returned", book.title)


class Library:

    def __init__(self):
        self.books = []

    def add_book(self, book):
        self.books.append(book)

    def show_books(self):
        print("\nLibrary Books:")
        for book in self.books:
            book.display_info()


# Creating books
b1 = Book("Python Basics", "John")
b2 = Book("Machine Learning", "Andrew Ng")

# Creating library
library = Library()

library.add_book(b1)
library.add_book(b2)

library.show_books()

# Creating users
s1 = Student("Vinod")
t1 = Teacher("Dr. Sharma")

# Borrow book
s1.borrow_book(b1)

library.show_books()

# Return book
s1.return_book(b1)

library.show_books()
````
