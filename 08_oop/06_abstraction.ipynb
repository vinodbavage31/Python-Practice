# Object-Oriented Programming: Abstraction (Python)

Abstraction is a key OOP concept that **focuses on showing only the essential features** of an object while **hiding the complex implementation details**.

---

### Concept

Think of it like using a **car**:

- When you drive a car:
  - Press accelerator → car moves  
  - Press brake → car stops  

- You **don’t need to know how the engine works internally**.  

The hidden complexity inside the car represents **abstraction**.

---

### Abstraction in Python

Python provides the **`abc` module** (Abstract Base Class) to implement abstraction.

- Abstract classes **cannot be instantiated directly**  
- They define **abstract methods** that **must be implemented by child classes**

---
### Question: Shape Area System (Using Abstraction)

#### Problem Description
Create a system that **calculates the area of different shapes**.

The **base class `Shape`** should **not implement the area calculation itself**.  
It should only **force child classes** to implement it.

---

#### Requirements

**Abstract Class → `Shape`**

**Abstract Method**
- `calculate_area()` → Must be implemented by child classes

**Child Classes**
- `Rectangle`  
- `Circle`

Each child class should **provide its own implementation** of `calculate_area()`.

---

#### Concepts Tested
- Abstraction  
- Abstract classes (`abc` module)  
- Enforcing method implementation in child classes

```python
from abc import ABC, abstractmethod


class Shape(ABC):

    @abstractmethod
    def calculate_area(self):
        pass


class Rectangle(Shape):

    def __init__(self, length, width):
        self.length = length
        self.width = width

    def calculate_area(self):
        area = self.length * self.width
        print("Rectangle Area:", area)


class Circle(Shape):

    def __init__(self, radius):
        self.radius = radius

    def calculate_area(self):
        area = 3.14 * self.radius * self.radius
        print("Circle Area:", area)


r = Rectangle(10, 5)
c = Circle(7)

r.calculate_area()
c.calculate_area()
```
