### Question 1: Product Inventory System

#### Problem Description
You are building a simple inventory system for a store.

Each product should store:
- Product name
- Price
- Quantity in stock

The system should allow:
- Displaying product details
- Updating stock when items are sold
- Adding stock when new items arrive

#### Requirements

Create a class called `Product`.

**Attributes**
- `name`
- `price`
- `quantity`

**Methods**
- `display_product()` → Prints product details
- `sell_product(amount)` → Decreases quantity
- `restock(amount)` → Increases quantity

#### Concepts Tested
- Constructor
- Object state update
- Instance methods

```python
class Product:

    def __init__(self, name, price, quantity):
        self.name = name
        self.price = price
        self.quantity = quantity

    def display_product(self):
        print("Product:", self.name)
        print("Price:", self.price)
        print("Quantity:", self.quantity)
        print("-------------------")

    def sell_product(self, amount):
        if amount > self.quantity:
            print("Not enough stock")
        else:
            self.quantity -= amount
            print(amount, "items sold")

    def restock(self, amount):
        self.quantity += amount
        print(amount, "items added to stock")


product1 = Product("Laptop", 60000, 10)

product1.display_product()

product1.sell_product(3)

product1.display_product()

product1.restock(5)

product1.display_product()
```

### Question 2: Simple Calculator Using Class

#### Problem Description
Create a calculator class that can perform basic operations.

The calculator should be able to:
- Add numbers
- Subtract numbers
- Multiply numbers
- Divide numbers

#### Requirements

Create a class called `Calculator`.

**Methods**
- `add(a, b)`
- `subtract(a, b)`
- `multiply(a, b)`
- `divide(a, b)`

#### Concepts Tested
- Methods inside classes
- Using objects to perform actions

```python
class Calculator:

    def add(self, a, b):
        print("Addition:", a + b)

    def subtract(self, a, b):
        print("Subtraction:", a - b)

    def multiply(self, a, b):
        print("Multiplication:", a * b)

    def divide(self, a, b):
        if b == 0:
            print("Cannot divide by zero")
        else:
            print("Division:", a / b)


calc = Calculator()

# taking input from user
a = float(input("Enter first number: "))
b = float(input("Enter second number: "))

calc.add(a, b)
calc.subtract(a, b)
calc.multiply(a, b)
calc.divide(a, b)
```

### Question 3: Mobile Phone System

#### Problem Description
Create a `MobilePhone` class that stores phone information.

Each phone should have:
- Brand
- Battery percentage

The phone should support:
- Making calls (reduces battery)
- Charging the phone
- Checking battery level

#### Requirements

Create a class called `MobilePhone`.

**Attributes**
- `brand`
- `battery`

**Methods**
- `make_call()` → Reduces battery by 10%
- `charge()` → Increases battery
- `check_battery()` → Displays the current battery level

```python
class MobilePhone:

    def __init__(self, brand, battery):
        self.brand = brand
        self.battery = battery

    def make_call(self):
        if self.battery >= 10:
            self.battery -= 10
            print("Call made successfully")
        else:
            print("Battery too low")

    def charge(self, amount):
        self.battery += amount
        if self.battery > 100:
            self.battery = 100
        print("Phone charged")

    def check_battery(self):
        print("Battery level:", self.battery, "%")


phone1 = MobilePhone("Samsung", 80)

phone1.check_battery()

phone1.make_call()

phone1.check_battery()

phone1.charge(15)

phone1.check_battery()
```

### Project Overview

### Problem Description

Create a system to **manage students**.

Each student should have:
- Name
- Age
- Marks

The system should allow:
- Displaying student details
- Updating marks
- Checking whether the student **passed or failed**

---

### Pass Condition
- `marks >= 40` → **Pass**  
- `marks < 40` → **Fail**

---

### Requirements

Create a class called `Student`.

**Attributes**
- `name`
- `age`
- `marks`

**Methods**
- `display_info()` → Displays student details  
- `update_marks()` → Updates the student's marks  
- `check_result()` → Checks whether the student passed or failed

```python
class Student:

    def __init__(self, name, age, marks):
        self.name = name
        self.age = age
        self.marks = marks

    def display_info(self):
        print("Name:", self.name)
        print("Age:", self.age)
        print("Marks:", self.marks)
        print("------------------")

    def update_marks(self, new_marks):
        self.marks = new_marks
        print("Marks updated successfully")

    def check_result(self):
        if self.marks >= 40:
            print(self.name, "has PASSED")
        else:
            print(self.name, "has FAILED")


# Creating objects
s1 = Student("Vinod", 21, 75)
s2 = Student("Bhavi", 22, 32)

# Display info
s1.display_info()
s2.display_info()

# Check results
s1.check_result()
s2.check_result()

# Update marks
s2.update_marks(50)

# Check again
s2.check_result()+
```
